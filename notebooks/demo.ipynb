# Quantitative Smart Beta Demo

本notebook演示如何使用Agnostic Risk Parity方法构建Smart Beta投资组合

```python
import sys
sys.path.insert(0, '../src')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from data_generator import DataGenerator
from smartbeta import SmartBeta
from methods import AgnosticRiskParity, RiskParity, EqualWeight
```

## 1. 生成模拟数据

```python
# 生成模拟因子数据
dg = DataGenerator(n_stocks=100, n_days=252*3)
factor_returns, factor_exposure = dg.generate()
print(f"因子收益: {factor_returns.shape}")
print(f"因子暴露: {factor_exposure.shape}")
```

## 2. 构建Smart Beta组合

```python
# 使用Agnostic Risk Parity
arp = AgnosticRiskParity()
weights = arp.fit(factor_returns)
print("Agnostic Risk Parity 权重:")
print(weights.head(10))
```

## 3. 绩效分析

```python
from metrics import calculate_sharpe, calculate_max_drawdown, calculate_returns

# 计算组合收益
portfolio_returns = calculate_returns(factor_returns, weights)

# 计算绩效指标
sharpe = calculate_sharpe(portfolio_returns)
mdd = calculate_max_drawdown(portfolio_returns)

print(f"夏普比率: {sharpe:.2f}")
print(f"最大回撤: {mdd:.2%}")
```

```python
# 可视化
plt.figure(figsize=(12, 6))
plt.plot((1 + portfolio_returns).cumprod())
plt.title('Portfolio Cumulative Returns')
plt.xlabel('Days')
plt.ylabel('Cumulative Return')
plt.show()
```
